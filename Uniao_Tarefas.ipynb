```python
import glob
import pandas as pd
import io
import gspread
```

```python
path = r'/content/drive/Shareddrives/Jornada do Cliente - SDV/Sustentação/Tarefas/'
```

```python
filenames = glob.glob(path + "/*.xlsx")
```

```python
print(filenames)
```

```python
lista_arquivos = []

for i in filenames:
  lista_arquivos.append(pd.read_excel(i))
```

```python
df_full = pd.concat(lista_arquivos, ignore_index=True)
```

```python
df_full.head(6000)
```

```python
dados = pd.DataFrame(df_full)


df_full = '/content/drive/MyDrive/Youtrack_full.xlsx'

dados.to_excel(df_full, index=False)
```
